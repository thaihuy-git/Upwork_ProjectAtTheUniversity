```python
from datetime import datetime
import random
import pandas as pd
import numpy as np
from faker import Faker
from faker.providers import DynamicProvider, BaseProvider

fake = Faker()

#Create providers
class JobCateProvider(BaseProvider):
    jc_list = [5, 6, 10, 12]
    def jobCate(self) -> int:
        return self.random_elements(self.jc_list, length=1, use_weighting=True)[0]
    
fake.add_provider(JobCateProvider)
#---Experience Level Provider
class ExperienceLevelProvider(BaseProvider):
    el_list = ['EL', 'EX', 'IT', 'ML']
    def ExLevel(self) -> str:
        return self.random_elements(self.el_list, length=1, use_weighting=True)[0]
    
fake.add_provider(ExperienceLevelProvider)



#---Job Duration
class JobDurationProvider(BaseProvider):
    jd = ['1 to 3 months', '3 to 6 months', 'Less than 1 month', 'More than 6 months']
    def JobDuration(self) -> str:
        return self.random_elements(self.jd, length=1, use_weighting=True)[0]
    
fake.add_provider(JobDurationProvider)

#---Working Hour
class WorkingHourProvider(BaseProvider):
    wh = ['Hours to be determined', 'Less than 30 hrs/week', '30+ hrs/week', 'not_sure']
    def WorkingHour(self) -> str:
        return self.random_elements(self.wh, length=1, use_weighting=True)[0]
    
fake.add_provider(WorkingHourProvider)

#---Applicant Num
class ApplicantNumProvider(BaseProvider):
    apNum = ['10 to 15', '15 to 20', '20 to 50', '5 to 10', '50+', 'Less than 5']
    def ApplicantNum(self) -> str:
        return self.random_elements(self.apNum, length=1, use_weighting=True)[0]
    
fake.add_provider(ApplicantNumProvider)

#---Payment Situation
class PaymentSituationProvider(BaseProvider):
    paySi = ['Payment unverified', 'Payment verified']
    def PaymentSituation(self) -> str:
        return self.random_elements(self.paySi, length=1, use_weighting=True)[0]
    
fake.add_provider(PaymentSituationProvider)


#---Enterprise Client
class EnterpriseClientProvider(BaseProvider):
    enClient = ['Enterprise', 'Small Business & Individual']
    def EnterpriseClient(self) -> str:
        return self.random_elements(self.enClient, length=1, use_weighting=True)[0]
    
fake.add_provider(EnterpriseClientProvider)

#---Connect Num
class ConnectNumProvider(BaseProvider):
    conNum = [2, 4, 8, 12, 16]
    def ConnectNum(self) -> int:
        return self.random_elements(self.conNum, length=1, use_weighting=True)[0]
    
fake.add_provider(EnterpriseClientProvider)

#---Rating
class RatingProvider(BaseProvider):
    rating = [1, 2, 3, 3.5, 3.7, 4, 4.3, 4.5, 4.7, 5, None, None, None, None, None, None, None]
    def Rating(self) -> int:
        return self.random_elements(self.rating, length=1, use_weighting=True)[0]
    
fake.add_provider(RatingProvider)

#---Payment Type
class PaymentTypeProvider(BaseProvider):
    payType = ['Fixed-price', 'Hourly']
    def PaymentType(self) -> str:
        return self.random_elements(self.payType, length=1, use_weighting=True)[0]
    
fake.add_provider(PaymentTypeProvider)

#---Keyword
df_jk = pd.read_csv("Data/JobCate_Keyword.csv")
def get_JobCate_Keywords(jc_id):
    jk_list = list(df_jk.query("JobCateID == " + str(jc_id))["Keyword"])
    return jk_list

class KeywordProvider(BaseProvider):
    def Keyword(self, jCateID) -> str:
        kWord = get_JobCate_Keywords(jCateID)
        return self.random_elements(kWord, length=1, use_weighting=True)[0]
    
fake.add_provider(KeywordProvider)

#---Skill: make sure that for each job post, the skills are not duplicated
df_js = pd.read_csv("Data/JobCate_Skills.csv")
def get_JobCate_Skill(jc_id, notList):
    js_list = list(df_js.query("JobCateID == " + str(jc_id) + "& SkillID not in " + str(notList))["SkillID"])
    return js_list

class SkillProvider(BaseProvider):
    def Skill(self, jCateID, notInList) -> int:
        sk = get_JobCate_Skill(jCateID, notInList)
        return self.random_elements(sk, length=1, use_weighting=True)[0]
    
fake.add_provider(SkillProvider)

#---Highlight
df_hl = pd.read_csv("Data/JobCate_Highlight.csv")
def get_Highlight(jc_id):
    hl_list = list(df_hl.query("JobCateID == " + str(jc_id))["Highlight"])
    return hl_list

class HighlightProvider(BaseProvider):
    def Highlight(self, jCateID) -> str:
        hlight = get_Highlight(jCateID)
        return self.random_elements(hlight, length=1, use_weighting=True)[0]
    
fake.add_provider(HighlightProvider)

#---Job Title
df_jt_jd = pd.read_csv("Data/JobTitle_JobDescription.csv")
def get_JobTitle(jc_id):
    jt_list = list(df_jt_jd.query("JobCateID == " + str(jc_id))["JobTitle"])
    return jt_list

class JobTitleProvider(BaseProvider):
    def pJobTitle(self, jCateID) -> str:
        jTitle = get_JobTitle(jCateID)
        return self.random_elements(jTitle, length=1, use_weighting=True)[0]
    
fake.add_provider(JobTitleProvider)

#---Job Description
def get_JobDescription(jc_id):
    jd_list = list(df_jt_jd.query("JobCateID == " + str(jc_id))["JobDescription"])
    return jd_list

class JobDescriptionProvider(BaseProvider):
    def pJobDescription(self, jCateID) -> str:
        jDescription = get_JobDescription(jCateID)
        return self.random_elements(jDescription, length=1, use_weighting=True)[0]
    
fake.add_provider(JobDescriptionProvider)

#---Spent
df_spent = pd.read_csv("Data/Spent.csv")
class SpentProvider(BaseProvider):
    def pSpent(self) -> float:
        spent_list = list(df_spent["Spent"])
        return self.random_elements(spent_list, length=1, use_weighting=True)[0]
    
fake.add_provider(SpentProvider)

#---JobCost
df_jcost = pd.read_csv("Data/JobCate_JobCost.csv")
def get_JobCost(jc_id):
    jcost_list = list(df_jcost.query("JobCateID == " + str(jc_id))["JobCost"])
    return jcost_list

class JobCostProvider(BaseProvider):
    def pJobCost(self, jCateID) -> float:
        jCost = get_JobCost(jCateID)
        return self.random_elements(jCost, length=1, use_weighting=True)[0]
    
fake.add_provider(JobCostProvider)

#---HourlyStartRate
df_HourlyRate = pd.read_csv("Data/JobCate_HourlyRate.csv")
def get_StartRate(jc_id):
    startRate_list = list(df_HourlyRate.query("JobCateID == " + str(jc_id))["StartRate"])
    return startRate_list

class StartRateProvider(BaseProvider):
    def StartRate(self, jCateID) -> float:
        sRate = get_StartRate(jCateID)
        return self.random_elements(sRate, length=1, use_weighting=True)[0]
    
fake.add_provider(StartRateProvider)

def get_EndRate(jc_id, paramStartRate):
    endRate_list = list(df_HourlyRate.query("JobCateID == " + str(jc_id) + " & EndRate > " + str(paramStartRate))["EndRate"])
    return endRate_list

class EndRateProvider(BaseProvider):
    def EndRate(self, jCateID, paStartRate) -> float:
        eRate = get_EndRate(jCateID, paStartRate)
        return self.random_elements(eRate, length=1, use_weighting=True)[0]
    
fake.add_provider(EndRateProvider)



#---Country list
df_ct = pd.read_csv("Data/Country.csv")
ct_list = list(df_ct["CountryISOCode"])

class CountryProvider(BaseProvider):
    def pCountry(self) -> str:
        return self.random_elements(ct_list, length=1, use_weighting=True)[0]

fake.add_provider(CountryProvider)
#=====================GENERATE DUMMY DATA
Faker.seed(42)
def generate_dummy_data(num, startDate, endDate):
    df_dummy = pd.DataFrame()
    for i in range(1, num+1):
        jobCateID = fake.jobCate()
        numOfSkill = random.randint(1,9)
        vSkill01 = fake.Skill(jobCateID, [])
        vSkill02 = None
        vSkill03 = None
        vSkill04 = None
        vSkill05 = None
        vSkill06 = None
        vSkill07 = None
        vSkill08 = None
        vSkill09 = None
        if(numOfSkill > 1):
            currentSkillList = [vSkill01]
            vSkill02 = fake.Skill(jobCateID, currentSkillList)
        if(numOfSkill > 2):
            currentSkillList.append(vSkill02)
            vSkill03 = fake.Skill(jobCateID, currentSkillList)
        if(numOfSkill > 3):
            currentSkillList.append(vSkill03)
            vSkill04 = fake.Skill(jobCateID, currentSkillList)
        if(numOfSkill > 4):
            currentSkillList.append(vSkill04)
            vSkill05 = fake.Skill(jobCateID, currentSkillList)
        if(numOfSkill > 5):
            currentSkillList.append(vSkill05)
            vSkill06 = fake.Skill(jobCateID, currentSkillList)
        if(numOfSkill > 6):
            currentSkillList.append(vSkill06)
            vSkill07 = fake.Skill(jobCateID, currentSkillList)
        if(numOfSkill > 7):
            currentSkillList.append(vSkill07)
            vSkill08 = fake.Skill(jobCateID, currentSkillList)
        if(numOfSkill > 8):
            currentSkillList.append(vSkill08)
            vSkill09 = fake.Skill(jobCateID, currentSkillList)
        
        df_dummy.loc[i, "jobCategoryID"] = jobCateID
        df_dummy.loc[i, "JobTitle"] = fake.pJobTitle(jobCateID)
        df_dummy.loc[i, "ExperienceLevelID"] = fake.ExLevel()
        df_dummy.loc[i, "SearchKeyword"] = fake.Keyword(jobCateID)
        df_dummy.loc[i, "PostedDate"] = fake.date_between(start_date=datetime.strptime(startDate,'%d/%m/%Y'), end_date=datetime.strptime(endDate,'%d/%m/%Y'))
        df_dummy.loc[i, "JobDurationID"] = fake.JobDuration()
        df_dummy.loc[i, "WorkingHourID"] = fake.WorkingHour()
        df_dummy.loc[i, "Highlight"] = fake.Highlight(jobCateID)
        df_dummy.loc[i, "JobDescription"] = fake.pJobDescription(jobCateID)
        df_dummy.loc[i, "Skill01"] = vSkill01
        df_dummy.loc[i, "Skill02"] = vSkill02
        df_dummy.loc[i, "Skill03"] = vSkill03
        df_dummy.loc[i, "Skill04"] = vSkill04
        df_dummy.loc[i, "Skill05"] = vSkill05
        df_dummy.loc[i, "Skill06"] = vSkill06
        df_dummy.loc[i, "Skill07"] = vSkill07
        df_dummy.loc[i, "Skill08"] = vSkill08
        df_dummy.loc[i, "Skill09"] = vSkill09
        df_dummy.loc[i, "ApplicantsNum"] = fake.ApplicantNum()
        df_dummy.loc[i, "PaymentSituation"] = fake.PaymentSituation()
        df_dummy.loc[i, "EnterpriseClient"] = fake.EnterpriseClient()
        df_dummy.loc[i, "Spent"] = fake.pSpent()
        df_dummy.loc[i, "CountryISOCode"] = fake.pCountry()
        df_dummy.loc[i, "Rating"] = fake.Rating()
        df_dummy.loc[i, "FeedbackNum"] = random.randint(0,100)
        df_dummy.loc[i, "PaymentType"] = fake.PaymentType()
        if(df_dummy.loc[i, "PaymentType"] == 'Fixed-price'):
            df_dummy.loc[i, "JobCost"] = fake.pJobCost(jobCateID)
        else:
            df_dummy.loc[i, "JobCost"] = None
            df_dummy.loc[i, "HourlyStartRate"] = fake.StartRate(jobCateID)
            df_dummy.loc[i, "HourlyEndRate"] = fake.EndRate(jobCateID, df_dummy.loc[i, "HourlyStartRate"])
            if(pd.isnull(df_dummy.loc[i, "HourlyEndRate"])):
                df_dummy.loc[i, "HourlyEndRate"] = df_dummy.loc[i, "HourlyStartRate"]
        
    
    return df_dummy

```

```python
df_2022_1 = generate_dummy_data(100000,'01/01/2022','31/12/2022')
df_2022_2 = generate_dummy_data(100000,'01/01/2022','31/12/2022')
df_2022_3 = generate_dummy_data(12651,'01/01/2022','31/12/2022')
df_2022_3
```

```python
df_2022 = df_2022_1
df_2022 = df_2022.append(df_2022_2)
df_2022 = df_2022.append(df_2022_3)
df_2022
```
