------------------------------CLEANING DATA---------------------------------------

###-------by Python
1. Remove unnecessary columns
2. Remove Job_Title is NULL
3. Remove Posted_from is NULL
4. Remove Client_Country is NULL
5. Update Spent is NULL --> 0
6. Split Time_Limitation into 2 columns: Time_Limitation and Hrs_Per_Week
7. Convert Posted_from to Short Date data type as a now column (Posted_Date), then, delete Posted_from
8. Update Feedback_Num that is null to 0
9. Remove "$" sign in Hourly_Rate, Job_Cost
10. Convert Job_Cost, Start_Rate, End_Rate to Number data type

###-------by SQL
11. Categorize jobs
12. Update EX_level_demand is NULL --> Intermediate, then categorizing Experience Level list
13. Standardize Country list
14. Categorize Skill list
15. Update Start_Rate and End_Rate with the same value if Hourly_Rate is just one value
16. Payment_Type = “Hourly” but Hour_Rate is empty --> update to the average hourly rate of job category
17. Payment_Type = “Fixed-price” but Job_Cost is empty --> update to the average of job category
18. The rating of clients:
    “Rating” is true if “Feedback_Num” is more than 10.
19. Update EnterpriseClient is NULL --> "Small Business & Individual"
21. Create the additional columns: JobFromHighRating, RatingGroup, Client, NumberOfApplicants

```python
#1. Remove unnecessary columns
new_df = df.drop({"Job_URL", "Category1_URL_search", "Category2_URL_search", "Category3_URL_search", "Category4_URL_search", "Category5_URL_search", "Category6_URL_search", "Category7_URL_search", "Category8_URL_search", "Category9_URL_search"},axis='columns')
```

```python
#change "Job Title" to "Job_Title"
new_df.rename(columns={"Job Title": "Job_Title"}, inplace=True)

#Change "Spent($)" to "Spent"
new_df.rename(columns={"Spent($)":"Spent"}, inplace=True)
```

```python
#2. Remove Job_Title is NULL
new_df.dropna(subset=["Job_Title"], inplace=True)

#3. Remove Posted_from is NULL
new_df.dropna(subset=["Posted_from"], inplace=True)

#4. Remove Client_Country is NULL
new_df.dropna(subset=["Client_Country"], inplace=True)

#5. Update Spent is NULL --> 0
values = {"EX_level_demand":"Intermediate", "Spent":0}
new_df2 = new_df.fillna(value=values)
new_df2["EX_level_demand"].value_counts(dropna=False)
new_df2["Spent"].value_counts(dropna=False)
```

```python
#6. Split Time_Limitation into 2 columns: Time_Limitation and Hrs_Per_Week
new_df2[["Time_Limitation2","Hrs_Per_Week"]] = new_df2["Time_Limitation"].str.split(', ', expand=True)
# Drop "Time_Limitation" column
# Rename "Time_Limitation2" --> "Time_Limitation"
new_df2 = new_df2.drop({"Time_Limitation"}, axis=1)
new_df2.rename(columns={"Time_Limitation2":"Time_Limitation"}, inplace=True)
```

```python
#7. Convert Posted_from to Short Date data type.
#----split Posted_from into 3 columns ("Posted_from_num", "Posted_from_unit", "Ago"), the purpose is to calculate the number of days.
#----Delete "Ago" column
#----Convert Posted_from_unit to number
#----Create and Calculate Posted_Date
new_df2[["Posted_from_num","Posted_from_unit", "Ago"]] = new_df2["Posted_from"].str.split(' ', expand=True)
new_df2 = new_df2.drop(["Ago"], axis=1)
new_df2 = new_df2.replace({"Posted_from_unit":{'day':1, 'days':1, 'hour':0, 'hours':0, 'minute':0, 'minutes':0, 'second':0, 'seconds':0}})
new_df2 = new_df2.astype({"Posted_from_num": int, "Posted_from_unit": int})
init_date = datetime.date(2024, 1, 2)
for i in new_df2.index:
    p_days = new_df2.loc[i, "Posted_from_num"] * new_df2.loc[i, "Posted_from_unit"]
    p_date = init_date + pd.Timedelta(days=p_days)
    new_df2.at[i, "Posted_Date"] = p_date
new_df2 = new_df2.drop(columns={"Posted_from_num", "Posted_from_unit", "Posted_from"})
```

```python
#8. Update Feedback_Num that is null to 0
new_df2["Feedback_Num"].value_counts(dropna=False)
```

```python
#9. Remove "$" sign in Hourly_Rate, Job_Cost, Start_Rate, End_Rate
new_df2["Hourly_Rate"] = new_df2["Hourly_Rate"].str.replace('$','')
new_df2["Job_Cost"] = new_df2["Job_Cost"].str.replace('$','')
new_df2["Start_rate"] = new_df2["Start_rate"].str.replace('$','')
new_df2["End_rate"] = new_df2["End_rate"].str.replace('$','')
```

```python
#10. Convert Job_Cost, Start_Rate, End_Rate to Number data type
new_df2["Hourly_Rate"] = new_df2["Hourly_Rate"].str.replace(' ','')
new_df2["Job_Cost"] = new_df2["Job_Cost"].str.replace(' ','')
new_df2["Start_rate"] = new_df2["Start_rate"].str.replace(' ','')
new_df2["End_rate"] = new_df2["End_rate"].str.replace(' ','')

new_df2["Hourly_Rate"] = new_df2["Hourly_Rate"].str.replace(',','')
new_df2["Job_Cost"] = new_df2["Job_Cost"].str.replace(',','')
new_df2["Start_rate"] = new_df2["Start_rate"].str.replace(',','')
new_df2["End_rate"] = new_df2["End_rate"].str.replace(',','')
new_df2 = new_df2.astype({"Job_Cost": float, "Start_rate": float, "End_rate": float})
```

```python
#Update Start_Rate and End_Rate
for i in new_df2.index:
    if str(new_df2.loc[i, "Hourly_Rate"]).find('-') == -1:
        if new_df2.loc[i, "Hourly_Rate"] is not None:
            new_df2.loc[i, "Start_rate"] = new_df2.loc[i, "Hourly_Rate"]
            new_df2.loc[i, "End_rate"] = new_df2.loc[i, "Hourly_Rate"]        
new_df2 = new_df2.drop(columns={"Hourly_Rate"}, axis=1)
```
